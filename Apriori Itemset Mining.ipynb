```python
def longest_line_length(lines):
    longest_count=0
    for line in lines:
        temp_count=0
        for item in line.split(";"):
            temp_count +=1
        if temp_count > longest_count:
            longest_count = temp_count
    return longest_count
```

```python
max_length = longest_line_length(lines)
print(max_length)
```

```python
def set_intersection(line, item_set):
    for item in item_set:
        if item not in line.split(";"):
            return False
    return True
first = set_intersection("Restaurants;Bars;American (New);Nightlife", frozenset(["American (New)","Nightlife"]))
second = set_intersection("Restaurants;Bars;American (New);Nightlife", frozenset(["American (New)","Nightlife","not"]))
third = set_intersection("Restaurants;Bars;Nightlife", frozenset(["American (New)","Nightlife"]))
assert(first == True)
assert(second == False)
assert(third == False)
onepattern
```

```python
def next_valid_candidates(frequent_items, database, min_count):
    final_dictionary = dict()
    temp_dictionary = dict()
    for item_set in frequent_items:
        for line in database:
            if set_intersection(line, item_set) and len(line.split(";")) > len(item_set):
                for i in line.split(";"):
                    if i not in item_set and frozenset([i]) in onepattern:
                        new_item = [i]
                        for i in item_set:
                            new_item.append(i)
                        new_item = frozenset(new_item)
                        if new_item in temp_dictionary:
                            temp_dictionary[new_item] += 1
                        else:
                            temp_dictionary[new_item] = 1
    for temp in temp_dictionary:
        if temp_dictionary[temp] >= min_count:
            final_dictionary[temp] = temp_dictionary[temp]
    return final_dictionary
```

```python
big_dictionary = dict()
for pattern in onepattern:
    big_dictionary[pattern] = onepattern[pattern]
k = 2
frequent_items = onepattern
while k <= max_length:
    frequent_items = next_valid_candidates(frequent_items, lines, line_count/100)
    for i in frequent_items:
        big_dictionary[i] = frequent_items[i]
    k += 1
```
